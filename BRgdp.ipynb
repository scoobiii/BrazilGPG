As previsões são feitas em um dataframe com uma coluna ds contendo as datas para as quais uma previsão deve ser feita. Você pode obter um dataframe adequado que se estenda no futuro por um número especificado de dias usando o método auxiliar Prophet.make_future_dataframe.
```python
# Python
m = Prophet()
m.fit(gdpBR)
```
