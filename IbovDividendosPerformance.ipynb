```python
ativos = [a+'.SA' for a in s_ativos.split(';')]
```

```python
ativos
```

```python
len(ativos)
```

```python
lista_d = []
for a in ativos:
  print('trabalhando no ativo = ',a)
  dados = yf.download(tickers=a,actions=True, period='Max')
  dp = (dados['Close'][-1]-dados['Open'][0])/dados['Open'][0]*100
  d = dados['Dividends']
  freq = d[d != 0].count()
  td = d.sum()
  lista_d.append([a, td, dp, freq])

df = pd.DataFrame(lista_d, columns=['Ativo','SomaDividendos','DeltaP','Freq'])
```

```python
df.sort_values(by='SomaDividendos',ascending=False)
```

```python
df.sort_values(by='DeltaP',ascending=False)
```
